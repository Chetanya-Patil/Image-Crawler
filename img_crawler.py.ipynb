```python
import requests
URL = "https://www.pexels.com/@hiteshchoudhary"
REQ = requests.get(URL).text
print(REQ)
```

```python
from bs4 import BeautifulSoup
soup = BeautifulSoup(REQ,'html.parser')
soup
```

```python
links = []
A = soup.select('img[src^="https://images.pexels.com/photos"]')

for images in A:
    links.append(images['src'])

print(A) #fully selected img elements
```

```python
for l in links:
    print(l)
```

```python
import os
os.mkdir('hitesh_photos')
i = 1
for index, img_link in enumerate(links):
    if i <= 30:
        img_data = requests.get(img_link).content
        with open("hitesh_photos/"+str(index+1)+'.jpg', 'wb+') as f:
            f.write(img_data)
        i += 1
    else:
        f.close()
        break
             
        
            
```
